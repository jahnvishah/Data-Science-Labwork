```python
corpus = df.copy()
```

```python
#convert the text to lowercase. 
for i in range(len(corpus)):
  corpus['text'][i] = corpus['text'][i].lower()
corpus.head()
```

```python
import re
for i in range(len(corpus)):
  corpus["text"][i] = re.sub(r'[^\w\s]', '', corpus["text"][i])
```

```python
corpus.head()
```

```python
import nltk
from nltk.corpus import stopwords
# nltk.download('stopwords')
print(len(stopwords.words('english')))
```

```python
corpus.head()
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

ps = PorterStemmer()
# stop_words = stopwords.words('english')
stop_words = pd.read_csv("stopwords.txt")
```

```python
nltk.download('punkt')
for i in range(len(corpus)):
  corpus['text'][i] = word_tokenize(corpus['text'][i])
  corpus["text"][i] = [ps.stem(w) for w in corpus["text"][i] if not w.lower() in stop_words]  

corpus.head()  
```

```python
from nltk.tokenize.treebank import TreebankWordDetokenizer
for i in range(len(corpus)):
  corpus["text"][i] = TreebankWordDetokenizer().detokenize(corpus["text"][i])
corpus.head()
```

```python
!pip install textmining3
import textmining
tdm = textmining.TermDocumentMatrix()
```

```python
for i in range(len(corpus)):
  tdm.add_doc(corpus["text"][i])
```

```python
tdm
```

```python
tdm.to_df(cutoff=0)
```

```python
threshold = (len(corpus)* 5 / 100)
stdm = tdm.to_df(cutoff=threshold)
stdm.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(stdm,corpus['spam'], test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
cart = DecisionTreeClassifier()
cart.fit(X_train, y_train)
y_pred = cart.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
spamRF = RandomForestClassifier()
spamRF.fit(X_train, y_train)
```

```python
y_pred = spamRF.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, cart.predict(X_test))
```

```python
roc_auc_score(y_test, spamRF.predict(X_test))
```
